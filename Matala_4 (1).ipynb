```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
url = "https://www.madlan.co.il/for-sale/ירושלים-ישראל"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36'}
response = requests.get(url, headers = headers)
soup = BeautifulSoup(response.content, 'html.parser')
if response.status_code == 200:
    print("success")
else:
    print(response.status_code)
```

```python
def get_info():
    url = "https://www.madlan.co.il/for-sale/%D7%99%D7%A8%D7%95%D7%A9%D7%9C%D7%99%D7%9D-%D7%99%D7%A9%D7%A8%D7%90%D7%9C"
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36'}
    response = requests.get(url, headers = headers)
    
    #Initialize DataFrame
    df = pd.DataFrame(columns=['City', 'type', 'room_number', 'Area', 'Street', 'number_in_street', 'city_area', 'price'])
    try:
        if not response.status_code == 200:
            return df
        soup = BeautifulSoup(response.content, 'html.parser')
        type_list = list()
        room_number_list = list()
        Area_list = list()
        Street_info = list()
        price_list = list()
        
# -------------------------------------------------------------  
        
        for type_ in soup.find_all('div', {'data-auto': 'property-class', 'class': 'css-1i00gl6 e1sx3tzs17'}):
            type_list.append(type_.get_text())
            
# -------------------------------------------------------------  

        for room in soup.find_all('div', {'data-auto': 'property-rooms', 'class': 'css-1i00gl6 e1sx3tzs17'}):
            room_number_list.append(room.get_text())
            
# -------------------------------------------------------------   

        for area in soup.find_all('div', {'data-auto': 'property-size', 'class': 'css-1i00gl6 e1sx3tzs17'}):
            Area_list.append(area.get_text())
            
# -------------------------------------------------------------

        for full_address in soup.find_all('div', {'data-auto': 'property-address', 'class': 'css-v4h09j e1sx3tzs16'}):
            Street_info.append(full_address.get_text()) 
            
        streets = []
        numbers = []
        city_areas = []

        for address in Street_info:
            # Extracting street using regex
            street_match = re.search(r'^([\w\s]+)', address)
            street = street_match.group(1).strip() if street_match else ""
            street = re.sub(r'\d+', '', street) if street else ""
            streets.append(street)

            # Extracting number in street using regex
            number_match = re.search(r'(\d+)', address)
            number = number_match.group(1).strip() if number_match else ""
            numbers.append(number)

            # Extracting city area using regex
            city_area_match = re.search(r'(?<=,)\s*([\w\s]+)$', address)
            city_area = city_area_match.group(1).strip() if city_area_match else ""
            city_areas.append(city_area)
            
# ------------------------------------------------------------- 

        for price in soup.find_all('div', {'data-auto': 'property-price'}):
            price_text = price.get_text()
            price_cut = price_text.strip()[:10]
            price_list.append(price_cut)
            
# -------------------------------------------------------------       
        
        df['City'] = "ירושלים"
        df['type'] = type_list
        df['room_number'] = room_number_list
        df['Area'] = Area_list
        df['Street'] = streets
        df['number_in_street'] = numbers
        df['city_area'] = city_areas
        df['price'] = price_list

        return df
    except:
        return df
```
